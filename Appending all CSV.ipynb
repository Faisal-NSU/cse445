```python
import pandas as pd
from csv import writer
import os
import glob

#returns the absolute path of the working directory where Python is currently running as a string str
path = os.getcwd() + '\\all responses' 

# make a list of all the csv files 
csv_files = glob.glob(os.path.join(path, "*.csv"))
```

```python
#loop through all csv files

# final dataset intially empty
newDF = []
# final dataset directory
finalDir = os.getcwd() + '\\final.csv' 

for f in csv_files:     
    # read the csv file
    df = pd.read_csv(f)
    
    # splitting file name to get country name
    csv_filename = f.split("\\")[-1]
    csv_filename = csv_filename[:len(csv_filename)-4]
    
    #drop Day column
    df.drop(["date"], axis=1, inplace = True) 
    
    #sum all the columns for this csv
    summed_list = df.sum(axis=0, numeric_only = False)
    summed_list = summed_list.tolist()
    summed_list.insert(0, csv_filename)
    newDF.append(summed_list)    

# converting the list to dataframe object and setting columns for it    
final = pd.DataFrame(newDF,columns=['Country','Corona', 'Corona virus', 'Covid cases', 'Face mask',
       'Hand sanitizer', 'Hand wash', 'Lockdown', 'N95', 'PPE',
       'Social distancing', 'Vaccine'])

# save the final dataframe to csv
final.to_csv(finalDir,index=False)
```
